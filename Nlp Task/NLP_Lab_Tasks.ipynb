```python

!pip install nltk pandas
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
import pandas as pd



text_data = [
    "The movie was fantastic and I loved every part of it about Egypt",
    "I hated the film, it was the worst I have ever seen",
    "The storyline was boring but the acting was brilliant",
    "An amazing movie with a great plot and incredible performances",
    "Egypt movie, I regret wasting my time on it",
    "The actors did a great job but the story lacked depth",
    "One of the best films I have seen in a long time, highly recommend it",
    "This film was just okay, not too bad but not great either",
    "Absolutely loved the movie, fantastic plot and wonderful cast",
    "The movie was disappointing, it did not live up to the hype"
]
```

```python
for sentence in text_data :
    print(word_tokenize(sentence))
```

```python

# 1. Tokenization
tokenized_data = [word_tokenize(sentence) for sentence in text_data]
for i, tokens in enumerate(tokenized_data):
    print(f"Sentence {i+1}:", tokens)
```

```python

# 2. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_data = [[word for word in tokens if word.lower() not in stop_words] for tokens in tokenized_data]
for i, tokens in enumerate(filtered_data):
    print(f"Sentence {i+1} (no stopwords):", tokens)
```

```python

# 3. Stemming and Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_data = [[stemmer.stem(word) for word in tokens] for tokens in filtered_data]
lemmatized_data = [[lemmatizer.lemmatize(word) for word in tokens] for tokens in filtered_data]

print("Stemming Example:", stemmed_data[0])
print("Lemmatization Example:", lemmatized_data[0])
```

```python

# 4. Part-of-Speech (POS) Tagging
pos_tagged_data = [pos_tag(tokens) for tokens in filtered_data]
for i, tagged in enumerate(pos_tagged_data):
    print(f"Sentence {i+1} POS Tags:", tagged)
```

```python

# 5. Named Entity Recognition (NER)
ner_data = [ne_chunk(tagged) for tagged in pos_tagged_data]
for i, tree in enumerate(ner_data):
    print(f"Sentence {i+1} NER Tree:")
    print(tree)
```

```python

# 6. Save Results to CSV
results = []
for i, sentence in enumerate(text_data):
    results.append({
        "Original": sentence,
        "Tokens": tokenized_data[i],
        "Filtered": filtered_data[i],
        "Stemmed": stemmed_data[i],
        "Lemmatized": lemmatized_data[i],
        "POS": pos_tagged_data[i]
    })

df = pd.DataFrame(results)
df.to_csv("nlp_lab_results.csv", index=False)
print("Results saved to nlp_lab_results.csv")
df.head()
```
